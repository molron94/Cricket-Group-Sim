```python
import random
```

```python
n_teams=6
fav_win_prob= 0.5
n_sims=100_000
sim_dict_unique=[]
for i in range(n_sims):
    sim_dict={}
    for j in range(n_teams):
        sim_dict[j]=0
        
    for j in range(n_teams):
        for k in range(j):
            if random.random()>fav_win_prob:
                sim_dict[j]+=1
                
            else:
                sim_dict[k]+=1
                
    val= list(set(list(sim_dict.values())))
    if len(val)==n_teams:
        sim_dict_unique.append(1)
        
    else:
        sim_dict_unique.append(0)
        
100*sum(sim_dict_unique)/n_sims
        
```
